```python
def B_matris_bul(m, n):
  B = np.zeros((m, n), int)
  for i in range(1, m + 1):
    for j in range(1, n + 1):
      if i == j:
        B[i - 1, j - 1] = i + j
      else:
       B[i - 1, j - 1] = i ** 2 + j
  return B
```

```python
B_matris_bul(3, 2)
```

```python
def M_matris_bul(m, n):
  M = np.zeros((m, n), int)
  for i in range(1, m + 1):
    for j in range(1, n + 1):
      if j - i == 1:
        M [i - 1, j - 1] = i
  return M  
```

```python
M_matris_bul(5, 5)
```

```python
M_matris_bul(7, 9)
```

```python
def N_harfi_olustur(n):
  N = np.zeros((n, n), int)
  for i in range(1, n + 1):
    for j in range(1, n + 1):
      if   i == j or j == 1 or j == n:
        N[i - 1, j - 1] = 1
  return N 
```

```python
N_harfi_olustur(5)
```

```python
N_harfi_olustur(10)
```

```python
def X_harfi_olustur(n):
  X = np.zeros((n, n), int)
  for i in range(1, n + 1):
    for j in range(1, n + 1):
      if  i == j or i + j == n + 1:
        X[i - 1, j - 1] = 1
  return X 
```

```python
X_harfi_olustur(10)
```

```python
def Z_harfi_olustur(n):
  Z = np.zeros((n, n), int)
  for i in range(1, n + 1):
    for j in range(1, n + 1):
      if  i + j == n + 1 or i == 1 or i == n:
        Z[i - 1, j - 1] = 1
  return Z 
```
